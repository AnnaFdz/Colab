## Inciso A (40 puntos)

Definir la clase `Adivinanza`, la cual es una abstracción de un **juego de adivinanzas**

### **Objetivo del juego**

Responder a una **pregunta** dentro de una cierta cantidad de intentos.

Si el **_adivinador_** responde correctamente dentro de 3 intentos entonces ganará, en caso contrario pierde.

### **¿En que consiste el juego?**

- Se establece una **pregunta**, una cantidad límite de intentos para responder y la **respuesta** a dicha pregunta. Para este ejercicio supondremos que la respuesta debe ser siempre una única palabra.
- Se proporciona un listado de **pistas** relacionadas con la adivinanza.
- Se realiza la **pregunta** a un jugador (al cual llamaremos el **adivinador**).
- El adivinador debe proponer una respuesta mientras cuente con algún intento.
- Se evalúa la respuesta propuesta por el adivinador:
    - Si es correcta, el advinador gana y se muestra un mensaje indicando esto.
    - Si es incorrecta y cuenta con algún intento, se debe mostrar el mensaje de respuesta incorrecta y la cantidad de intentos restantes. Luego el adivinador puede volver a proponer otra respuesta.
    - Si es incorrecta y solo tenía un intento, el adivinador pierde y se muestra un mensaje indicando esto.
- El adivinador puede solicitar una **pista** mientras tenga al menos un intento.

La clase a definir deberá contar con:

1. Un atributo que registre la **pregunta** realizada.
2. Un atributo que registre la **respuesta** de la adivinanza.
3. Un atributo que almacene el listado de **pistas** sobre la adivinanza.
4. Un atributo que registre la cantidad **intentos** restantes.

## Además, deberá contar con los siguientes métodos:

1. Un método que **inicialice** una adivinanza, indicando una **pregunta**, su **respuesta** y una lista de **pistas**.
2. Un método que muestre la **pregunta** de la adivinanza.
3. Un método que muestre una **pista** del listado de pistas. Al hacerlo, dicha pista debe ser eliminada del listado. Si se agotan las pistas y/o intentos debe mostrar un mensaje indicando esto.
4. Un método que permita **adivinar**, en el cual se indicará una posible solución a la adivinanza (una palabra que el adivinador propondrá). Devolverá `True` si adivinó la respuesta y `False` en caso contrario.
    - Si no tiene intentos debe mostrar un mensaje indicando que el juego ha terminado.
    - Si la respuesta es incorrecta, deberá decrementar la cantidad de **intentos** en uno y mostrar un mensaje indicando los intentos restantes. Además, si al realizar este decremento, la cantidad de intentos se vuelve cero `0` entonces debe mostrar un mensaje indicando que ha perdido el juego.
    - Si la respuesta es correcta, deberá mostrar un mensaje indicando que ganó el juego y cambiar la cantidad de intentos a `0`.

### Consideraciones

- Al iniciar un objeto de esta clase la cantidad de **intentos** es 3.

## Inciso B (15 puntos)

A partir de la clase implementada en el inciso anterior, simular una partida del juego de adivinanzas: 

1. Suponiendo que la Adivinanza propuesta tiene como:
    - **Pregunta**: `"Treinta caballos de color blanco. En una sierra colorada, que cuando cabalgan machacan y trituran pero luego descansan."`
    - **Respuesta**: `"dientes"`
    - **Pistas**: `["Son parte del cuerpo","Están compuestos por calcio y fósforo", "Puedes verlos al sonreir"]`
2. Mostrar la pregunta de la adivinanza
3. Mientras tengas intentos:
    - Intentar adivinar la respuesta. Si falla, pide una pista.

A continuación, mostramos algunos ejemplos:

```python
# Inciso A
class Adivinanza:

    def __init__(self, pregunta, respuesta, pistas):
        self.pregunta = pregunta
        self.respuesta = respuesta
        self.pistas = pistas
        self.intentos = 3
    
    def mostrar_pregunta(self):
        print(self.pregunta)
    
    def mostrar_pista(self):
        if len(self.pistas)>0 and self.intentos>0:
            print(self.pistas.pop())
        elif len(self.pistas)>0 and not self.intentos>0:
            print("El juego ha terminado")
        else:
            print("No te quedan más pistas")

    def adivinar(self,respuesta):
        if self.intentos>0:
            if self.respuesta == respuesta:
                print("Felicidades adivinaste! Has ganado!")
                self.intentos = 0
                return True
            else:
                self.intentos -= 1
                print(f"Te equivocaste. Te quedan {self.intentos} intentos")
                if self.intentos == 0:
                    print("Has perdido el juego")
                return False
        else:
            print("El juego ha terminado")
    
# Inciso B
pregunta = "Treinta caballos de color blanco. En una sierra colorada, que cuando cabalgan machacan y trituran pero luego descansan."
respuesta = "dientes"
pistas = ["Son parte del cuerpo","Están compuestos por calcio y fósforo", "Puedes verlos al sonreir"]
adivinanza = Adivinanza(pregunta,respuesta,pistas)
adivinanza.mostrar_pregunta()
while adivinanza.intentos>=1:
    mi_respuesta = input("Intente adivinar la respuesta: ")
    if not adivinanza.adivinar(mi_respuesta):
        adivinanza.mostrar_pista()
```

## Inciso B (15 puntos)

A partir de la clase implementada en el inciso anterior, simular la siguiente situación:

1. Se dispone de un dron cuya altura máxima es de 19 metros.
2. Hacer que el dron despegue.
3. Se le indicará al dron ascender hasta los 9 metros de la altura.
4. Posteriormente, se le indicará al dron que ascienda otros 12 metros.
5. Finalmente, indicarle al dron que descienda 21 metros.

A continuación, mostramos algunos ejemplos:

```python
# Inciso A
class Dron:

  def __init__(self, modelo, altura_maxima):
    self.modelo = modelo
    self.altura_maxima = altura_maxima
    self.altura_actual = 0

  def despegar(self):
    self.altura_actual = 1
    print(
      f"El dron {self.modelo} ha despegado. Se encuentra a 1 metro de altura")

  def aterrizar(self):
    self.altura_actual = 0
    print(
      f"El dron {self.modelo} ha aterrizado. Se encuentra a 0 metros de altura"
    )

  def ascender(self, metros):
    if self.altura_actual + metros <= self.altura_maxima:
      self.altura_actual += metros
      print(
        f"El dron {self.modelo} ha ascendido hasta los {self.altura_actual} metros."
      )
    else:
      metros = self.altura_actual + metros - self.altura_maxima
      self.altura_actual = self.altura_maxima
      print(
        f"El dron {self.modelo} ha ascendido a la altura máxima ({self.altura_maxima} metros). No es posible ascender {metros} metros más"
      )

  def descender(self, metros):
    if self.altura_actual - metros >= 0:
      self.altura_actual -= metros
      print(
        f"El dron {self.modelo} ha descendido hasta {self.altura_actual} metros"
      )
    else:
      metros = metros - self.altura_actual
      self.aterrizar()
      print(
        f"No es posible descender {metros} metros más, ya que el dron ha aterrizado"
      )


# Inciso B

modelo = input("Indique el modelo del dron: ")
am = int(input("Indique altura máxima del dron: "))
dron = Dron(modelo, am)
dron.despegar()
metros = int(
  input("Indique la cantidad de metros que el dron debe ascender: "))
dron.ascender(metros)
metros = int(
  input("Indique la cantidad de metros que el dron debe ascender: "))
dron.ascender(metros)
metros = int(
  input("Indique la cantidad de metros que el dron debe descender: "))
dron.descender(metros)
            
```

## Inciso B (15 puntos)

A partir de la clase implementada en el inciso anterior, simular la siguiente situación:

1. Se dispone de una cafetera con una capacidad para 5 tazas.
2. Se intenta servir 1 taza de café.
3. Se preparan 6 tazas de café.
4. Se sirven 3 tazas de café.
5. Se sirven otras 3 tazas de café.
6. Se revisa cuantas tazas de café le quedan a la cafetera:
    - Si queda alguna taza, entonces se sirve 1 taza.
    - Si no queda ninguna taza, entonces se rellena 1 taza y luego se sirve dicha taza.

```python
# Inciso A
class Cafetera:

  def __init__(self, capacidad_maxima):
    self.capacidad_maxima = capacidad_maxima
    self.cantidad_actual = 0

  def __len__(self):
    return self.cantidad_actual

  def hacer_cafe(self, tazas):
    if self.cantidad_actual + tazas <= self.capacidad_maxima:
      print(f"Haciendo {tazas} tazas de café")
      self.cantidad_actual += tazas
    else:
      print(
        f"Haciendo {self.capacidad_maxima-self.cantidad_actual} tazas de café")
      self.cantidad_actual = self.capacidad_maxima
      print(f"La cafetera esta llena")

  def servir_cafe(self, tazas):
    if tazas <= self.cantidad_actual:
      print(f"Sirviendo {tazas} tazas de café")
      self.cantidad_actual -= tazas
    else:
      print(f"Sirviendo {self.cantidad_actual} tazas de café")
      self.cantidad_actual = 0
      print(f"La cafetera esta vacía")


# Inciso B
tazas = int(input("Indique capacidad máxima tazas de la cafetera: "))
cafetera = Cafetera(tazas)
tazas = int(input("Indique la cantidad de tazas que desea hacer: "))
cafetera.hacer_cafe(tazas)
tazas = int(input("Indique la cantidad de tazas que desea hacer: "))
cafetera.hacer_cafe(tazas)
tazas = int(input("Indique la cantidad de tazas que desea servir: "))
cafetera.servir_cafe(tazas)
tazas = int(input("Indique la cantidad de tazas que desea servir: "))
cafetera.servir_cafe(tazas)
if len(cafetera) >= 1:
  tazas = int(input("Indique la cantidad de tazas que desea servir: "))
  cafetera.servir_cafe(tazas)
else:
  tazas = int(input("Indique la cantidad de tazas que desea hacer: "))
  cafetera.hacer_cafe(tazas)
  tazas = int(input("Indique la cantidad de tazas que desea servir: "))
  cafetera.servir_cafe(tazas)
```
